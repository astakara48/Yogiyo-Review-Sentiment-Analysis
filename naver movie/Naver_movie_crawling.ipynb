```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import myslack
import datetime as dt
import pickle
```

```python
driver = webdriver.Chrome()
```

```python
def naver_movie_1_1000():
    df = pd.DataFrame(columns=['Author','Date','Movie_Title','Review','Point'])
    for i in range(1,1001):
        try:
            url = 'https://movie.naver.com/movie/point/af/list.nhn?&page={}'.format(i)
            response = requests.get(url)
            dom = BeautifulSoup(response.content, 'html.parser')
            rows = dom.select('.list_netizen tbody tr')
            for row in rows:
                df.loc[len(df)] = { 
                    'Author':row.select_one('.num .author').text.split('*')[0],
                    'Date':str(row).split('<br/>')[2].split('<')[0].replace('.','/'),
                    'Movie_Title':row.select_one('.title .movie').text,
                    'Review':row.select_one('.title').text.split('\r')[0].split('\n')[2].strip(),
                    'Point':row.select_one('.point').text,
                }
                
        except AttributeError:
            pass
        
        print('page {} finish'.format(i))
    print('All work done!')
    
    return df
    
naver_movie_all_df = naver_movie_1_1000()
```

```python
myslack.send_slack('finish!')
```
